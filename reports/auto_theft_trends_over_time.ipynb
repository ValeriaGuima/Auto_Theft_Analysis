```python
#Import required Libraries
import pandas as pd

#read cleaned_data1.csv
file_path = '..\data\processed\cleaned_data1.csv'
df= pd.read_csv(file_path)

print(df.head)
```

```python
#import plotly express library for visualization
import plotly.express as px

#Generate histogram to show incidents reported in each year to see yearly incident trends. Dataset used was latest updated July 23rd 2024
fig_incident_count = px.histogram(df,x="OCC_YEAR", 
                                  labels={
                                      "OCC_YEAR":"Year",
                                      "count":"Count of Occurences"
                                  },
                                  title='Auto Theft Reported Incidents by Year')
fig_incident_count.show()

# Histogram shows that there is a steady increase in auto thefts over time. 
# While 2022 and 2023 may have been outliers, 2024 incidences is trending above 2021 and prior years.
# This indicates that without intervention, the environment has shifted to likely expect increased reported auto theft occurances since 2021.  
```
